# 1.3 二维数据的数字特征及相关分析

## 1.3.1 二维数据的数字特征及相关系数

二维总体$(X,Y)^{T}$，其样本数据：$(x_{i},y_{i})^{T}$，$i=1,2 ...,n$

### (1)二维数据的均值

样本均值$\overline{x}=\frac{1}{n}\sum_{i=1}^n x_{i} ,\overline{y}=\frac{1}{n}\sum_{i=1}^n y_{i}$

样本方差$s_{xx}=\frac{1}{n-1}\sum_{i=1}^n (x_{i}-\overline{x})^{2} ,s_{yy}=\frac{1}{n-1}\sum_{i=1}^n (y_{i}-\overline{y})^{2}$

### (2)样本的协方差

$$s_{xy}=\frac{1}{n-1}\sum_{i=1}^n(x_{i}-\overline{x})(y_{i}-\overline{y})$$

很显然，均值描述的是样本集合的中间点，它告诉我们的信息是很有限的，而标准差给我们描述的则是样本集合的各个样本点到均值的距离之平均。以这两个集合为例，[0，8，12，20]和[8，9，11，12]，两个集合的均值都是10，但显然两个集合差别是很大的，计算两者的标准差，前者是8.3，后者是1.8，显然后者较为集中，故其标准差小一些，标准差描述的就是这种“散布度”。之所以除以n-1而不是除以n，是因为这样能使我们以较小的样本集更好的逼近总体的标准差，即统计上所谓的“无偏估计”。而方差则仅仅是标准差的平方。

协方差其意义在于：**度量各个维度偏离其均值的程度**。协方差的值如果为正值，则说明两者是正相关的(从协方差可以引出“相关系数”的定义)，结果为负值就说明负相关的，如果为0，也是就是统计上说的“相互独立”。

### (3)皮尔逊相关系数

两个变量之间的皮尔逊相关系数定义为两个变量之间的协方差和标准差的商：

$$R=\frac{S_{xy}}{\sqrt{S_{xx}}\sqrt{S_{yy}}}$$

用来**衡量两个数据集合是否在一条线上面**，它用来衡量定距变量间的线性关系，但相关系数只是刻画了X，Y之间的“线性”关系程度。换句话说，假如X与Y有其它的函数关系但非线性关系时，用相关系数来衡量是不合理的。皮尔逊相关系数的变化范围为-1到1。相关系数的绝对值越大，则表明X与Y相关度越高。当X与Y线性相关时，相关系数取值为1（正线性相关）或-1（负线性相关）。 系数的值为1意味着X和Y可以很好的由直线方程来描述，所有的数据点都很好的落在一条直线上，且$Y$随着$X$的增加而增加。

### (4)斯皮尔曼相关系数

$$q_{xy}=\frac{\sum_{i=1}^{n}(R_{i}-\overline{R})(S_{i}-\overline{S})}{\sqrt[]{\sum_{i=1}^{n}(R_{i}-\overline{R})^{2}}\sqrt[]{\sum_{i=1}^{n}(S_{i}-\overline{S})^{2}}}$$

斯皮尔曼相关可以看作是皮尔逊相关的非参数版本,能够适用皮尔逊相关的场合当然是优先使用皮尔逊相关，但是在有些场合，皮尔逊相关所需要的前提假设不能得到满足，比如要符合正态分布、样本容量要超过一定数量(比如30个)，这就可以考虑使用斯皮尔曼相关，比如说以下一些情况下：

1.如果你的数据展现的是非线性关系，或者不是正态分布的。

2.如果至少有一方数据是序数类型（ordinal）而非数值类型。比如说，如果数据的赋值为"第一、第二、第三、... "你就是在处理序数类型数据。更具具体一点的例子就是，比如说你考察两个球队在历年联赛中的战绩之间的关系，那么你得到的数据可能是这样的：A队在2010~2020年间的联赛排名为{1,2,4,5,...,2}, B队在2010 ~ 2020年间的联赛排名为{2,1,3,6,...,4}。这两个数据就是序数类型的数据，考察它们的相关性你使用皮尔逊相关系数就不妥当

3.如果数据中有明显的异常值（outliers）。与皮尔逊相关不同，斯皮尔曼相关对于异常值不太敏感，因为它基于排序位次进行计算，实际数值之间的差异大小对于计算结果没有直接影响

最后再来对比一下皮尔逊相关系数和斯皮尔曼相关系数：如果数据是连续的、正态分布的，并且是线性的，优先使用皮尔逊相关系数。当然，也可以使用斯皮尔曼，但效率没有皮尔逊高(因为要进行排序)；如果前面几个条件有一个不满足，则使用斯皮尔曼相关系数；皮尔逊相关系数适用于线性关系，而斯皮尔曼相关系数适用于单调关系，线性与单调的一个区别是，线性关系的斜率是固定的。如果数据看上去，既有点像线性关系，又有点像单调关系，那么使用斯皮尔曼相关系数；皮尔逊相关系数使用元数据进行计算的，而斯皮尔曼相关系数是基于秩计算的。

## 1.3.2多维数据的数字特征及相关矩阵

### 皮尔逊相关矩阵，斯皮尔曼相关矩阵

$$R=\begin{bmatrix}1&  r_{12}&  ...&  r_{1p}\\r_{21}&  1&  ...&  r_{2p}\\...&  ...&  ...&  ...\\r_{p1}&  r_{p2}&  ...&  1\\\end{bmatrix}\Rightarrow 样本数据的Pearson相关矩阵$$

$$r_{jk}=\frac{s_{jk}}{\sqrt{s_{jj}}\sqrt{s_{kk}}}=\frac{s_{jk}}{s_{j}s_{k}}\Rightarrow X的第j,k列数据的相关系数$$

注意：

1.Spearman相关矩阵适用于研究具有一般分布的维总体，且对异常值具有耐抗性,同样非负定的。

2.原数据的相关系数即为标准化数据观测矩阵的协方差阵。

$$Q=\begin{bmatrix}1&  q_{12}&  ...&  q_{1p}\\q_{21}&  1&  ...&  q_{2p}\\...&  ...&  ...&  ...\\q_{p1}&  q_{p2}&  ...&  1\\\end{bmatrix}\Rightarrow 多维数据的秩相关系数矩阵$$

$$q_{jk}=\frac{\sum_{n}^{i=1}(R_{ij}-\overline{R_{j}})(R_{ik}-\overline{R_{k}})}{\sqrt{\sum_{n}^{i=1}(R_{ij}-\overline{R_{j}})^{2}}\sqrt{\sum_{n}^{i=1}(R_{ik}-\overline{R_{k}})^{2}}}$$

注意：$q_{jk}$为$X_{j}$和$X_{k}$的Spearson（秩）相关系数，$j,k=1,2,...,p$;

$D=Diag(\sqrt{s_{11},}\sqrt{s_{22}},...,\sqrt{s_{pp}})$为对角阵，且$R=D^{-1}SD^{-1}$,$R$与$S$相似。

对$X_{j}$数据标准化,得标准化数据矩阵；每行为标准化后样品观测向量

$$S^{*}=\frac{1}{n-1}\sum_{n}^{i=1}X^{*}_{i}(X^{*}_{i})^{T}=\frac{1}{n-1}\sum_{n}^{i=1}(X^{*}_{i})^{T}X^{*}_{i}=R\Rightarrow 原数据的相关矩阵$$

注意:原数据的相关系数矩阵即为标准化数据协方差阵。


## 1.3.3总体的数字特征、相关矩阵及多维数据正态分布

### （1）p维总体均值向量

均值向量：多维随机样本均值向量亦称样本均值，多元分析的基本概念之一，是对多维随机样本取平均值所得的向量。
具体作法是对每个分量的抽样取平均值。

$$\mathbf{\mu} =(\mu_{1},\mu_{2},...,\mu_{p})^{T}$$
$$\mu_{i}=E(X_{i}),i=1,2,...,p$$

### （2）协方差矩阵、总体相关系数矩阵

协方差多了就是协方差矩阵，协方差也只能处理二维问题，那维数多了自然就需要计算多个协方差，比如n维的数据集就需要计算 $n! / ((n-2)!*2) $个协方差，那自然而然的我们会想到使用矩阵来组织这些数据。给出协方差矩阵的定义：

$$C_{n_{X}n}=(c_{i,j},c_{i,j}=cov(Dim_{i},Dim_{j}))$$

这个定义还是很容易理解的，我们可以举一个简单的三维的例子，假设数据集有三个维度，则协方差矩阵为


$$C=\begin{pmatrix}cov(x,x)&  cov(x,y)&  cov(x,z)\\cov(y,x)&  cov(y,y)&  cov(y,z)\\cov(z,x)&  cov(z,y)&  cov(z,z）\\\end{pmatrix}$$

协方差矩阵是一个对称的矩阵，而且对角线是各个维度上的方差。

总体相关系数矩阵：

$$\rho=\begin{bmatrix}1&  \rho_{12}&  ...&  \rho_{1p}\\\rho_{21}&  1&  ...&  \rho_{2p}\\...&  ...&  ...&  ...\\\rho_{p1}&  \rho_{p2}&  ...&  1\\\end{bmatrix}$$

注意：

（1）$\rho_{jk}=\frac{\sigma _{jk}}{\sqrt{\sigma _{jj}}\sqrt{\sigma _{kk}}}=\frac{\sigma _{jk}}{\sigma _{j}\sigma _{k}}$为$X_{j},X_{k}$的相关系数满足$\rho_{jj}=1，|\rho|\leq 1$；

（2）记$V=Diag(\sqrt{\sigma _{11},}\sqrt{\sigma _{22}},...,\sqrt{\sigma _{pp}})=Diag(\sigma _{1},\sigma _{2},...,\sigma _{n})$则$\rho = V^{-1}\sum V^{-1}$也是非负定矩阵

$当n充分大时，有 \mu \approx \overline{X} $   $\sum \approx S$   $\rho \approx R$

### (3)多维正态分布

二维正太分布：$(X_{1},X_{2}) \sim N(\mu _{1},\mu _{2},\sigma _{1}^{2},\sigma _{2}^{2},\rho )$

$$f(x)=\frac{1}{(2\pi )^{\frac{2}{2}}|\sum |^{\frac{1}{2}}}exp{-\frac{1}{2}[(x-\mu )^{T}\sum^{-1}(x-\mu )]}$$

p维正态分布：

$X=(X_{1},X_{2},...,X_{p})^{T} \sim N_{p}(\mu , \sim )$——p维正态分布，密度

$$f(x)=f(x_{1},x_{2},...,x_{p})=\frac{1}{(2\pi )\frac{p}{2}|\sum |\frac{1}{2}} exp {-\frac{1}{2}[(x-\mu )^{T}\sum ^{-1}(x-\mu)]}$$

$$\mu = E(X)=(E(X_{1}),...,E(X_{p})^{T}=(\mu _{1},\mu _{2},...,\mu _{p}))^{T}——均值向量$$
